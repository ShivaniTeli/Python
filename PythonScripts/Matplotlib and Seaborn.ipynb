```python
weekdays = [1,2,3,4,5,6,7]
temp = [45,36,21,68,33,52,40]
```

```python
plt.plot(weekdays,temp)

plt.xlabel('Days')
plt.ylabel('Temp')
plt.title('Weather Data')
plt.show()
```

```python
plt.plot(weekdays,temp)

plt.yticks(range(20,100,5))
plt.xlabel('Days')
plt.ylabel('Temp')
plt.title("Weather Data")

plt.show()
```

```python
plt.plot(weekdays,temp,c = 'r')

plt.xlabel('Days')
plt.ylabel('Temp')
plt.title('Weather Data')
plt.show()
```

```python
plt.plot(weekdays,temp,c = 'r',ls = '--')

plt.xlabel('Days')
plt.ylabel('Temp')
plt.title('Weather Data')
plt.show()
```

```python
plt.plot(weekdays,temp,c = 'r',ls='--',marker='o')

plt.xlabel('Days')
plt.ylabel('Temp')
plt.title('Weather Data')
plt.show()
```

```python
days=[1,2,3,4,5,6,7] # weekdays
max_t=[50,51,52,48,47,49,46] # max temp recorder at that day
min_t=[43,42,40,44,45,35,37] # min temp
avg_t=[45,48,48,46,40,42,41] # avg temp
```

```python
plt.plot(days,max_t,label = 'MAX',marker = 'v',ms=10)
plt.plot(days,min_t,label = 'MIN',marker = 'o',ms=10)
plt.plot(days,avg_t,label = 'AVG',marker = 's',ms=10)

plt.legend()
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather Data")

plt.show()
```

# Histogram : frequency of data within particular range

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
```

80 - 100 ---> Normal

100-125---> prediabetic

> 125 ----> diabetic

```python
plt.hist(blood_sugar,bins=[77,100,125,150])

plt.grid()
plt.show()
```

# Pie

```python
on_what = ['rent' , 'car' , 'internet' , 'food' , 'other']
monthly_expense = [1400 , 534,645,234,563]
```

```python
plt.pie(monthly_expense)

plt.show()
```

```python
plt.pie(monthly_expense,labels=on_what,autopct='%0.1f%%')

plt.show()
```

```python
plt.pie(monthly_expense,labels=on_what,autopct='%0.f%%',radius = 1.5)

plt.show()
```

```python
plt.pie(monthly_expense,labels=on_what,autopct='%0.f%%',radius = 1.5,explode = [0,0,0.5,0,0])

plt.show()
```
