# Normal Slicing: 
* using normal slicing you can only extract rows but can't extract on columns

1.zero indexing:
-start inclusive
-stop exclusive

2.custom indexing:
-start inclusive
-stop inclusive

# Loc:

* using loc function slicing can be performed on rows and columns
* only custom indexing allowed

1.custom indexing:
-start inclusive
-stop inclusive


# iloc:
* using normal slicing can be performed on rows and columns
* only zero indexing allowed

1.zero indexing:
-start inclusive
-stop exclusive

```python
pd.merge(df1,df2,on = 'city',how ='right')
```

```python
pd.merge(df1,df2,on = 'city',how ='outer')
```
