# Group by

```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/weather_by_cities_group_by.csv')

df
```

```python
df['city'].unique()
```

```python
df['event'].value_counts()
```

```python
df['temperature'].max()
```

```python
temp =df.groupby('city')

list(temp)
```

```python
temp.max(numeric_only = True)
```

```python
temp.max(numeric_only = True)['temperature']
```

```python
temp.mean(numeric_only = True)['windspeed']
```

```python
import seaborn as sns

df_titanic = sns.load_dataset('titanic')

df_titanic
```

```python
demo = df_titanic.groupby('sex')

list(demo)
```

```python
demo.sum(numeric_only = True)['survived']
```

```python
demo.count()['survived']
```

```python
(demo.sum(numeric_only = True)['survived'] / demo.count()['survived']) * 100
```
